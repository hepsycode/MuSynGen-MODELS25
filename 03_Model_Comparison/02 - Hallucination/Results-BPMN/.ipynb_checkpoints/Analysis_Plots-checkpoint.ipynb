```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
categories = ["Connecting Objects", "Flow Objects", "Swim Lanes", "Artifacts"]
llm_data = [
    [44, 46, 37, 68, 39, 38, 28, 30, 54, 58],  # Connecting Objects
    [10, 9, 8, 13, 9, 9, 10, 7, 13, 16],       # Flow Objects
    [2, 2, 3, 4, 3, 6, 4, 3, 2, 4],            # Swim Lanes
    [0, 0, 2, 1, 2, 4, 0, 0, 0, 0],            # Artifacts
]
human_data = [
    [2556, 2472, 2572, 3607, 729, 1270, 5006, 3691, 724, 5520],  # Connecting Objects
    [806, 757, 626, 987, 200, 367, 1592, 1534, 263, 1722],        # Flow Objects
    [0, 0, 221, 264, 22, 68, 312, 91, 57, 60],                    # Swim Lanes
    [0, 0, 224, 246, 37, 56, 21, 3, 0, 9],                        # Artifacts
]

# Combine data
examples = [f"Ex{i:02}" for i in range(1, 11)]
x_labels = [f"{ex}-{model}" for ex in examples for model in ["LLM", "Human"]]

# Combine LLM and Human data for normalization
llm_totals = np.sum(llm_data, axis=0)
human_totals = np.sum(human_data, axis=0)

llm_percentages = (np.array(llm_data) / llm_totals * 100).T  # Percentages for LLM
human_percentages = (np.array(human_data) / human_totals * 100).T  # Percentages for Human

# Combine percentages into a single array
combined_percentages = np.concatenate((llm_percentages, human_percentages), axis=0)

# Configure the plot
x = np.arange(len(x_labels))  # X positions
width = 0.8  # Width of the bars

fig, ax = plt.subplots(figsize=(15, 8))

# Bottom tracking for stacking bars
bottom = np.zeros_like(x, dtype=float)

# Add stacked bars for each category
for i, category in enumerate(categories):
    ax.bar(x, combined_percentages[:, i], width, label=category, bottom=bottom)
    bottom += combined_percentages[:, i]  # Update the bottom for the next stack

# Customize the plot
ax.set_xlabel("Examples (LLM and Human)")
ax.set_ylabel("Percentage")
ax.set_title("Percentage Stacked Histogram for LLM and Human")
ax.set_xticks(x)
ax.set_xticklabels(x_labels, rotation=45, ha="right")  # Rotate labels for better visibility
ax.legend(title="Categories")
ax.grid(axis="y", linestyle="--", alpha=0.7)

# Save the plot
file_name = "llm_human_percentage_stacked_bar.png"
plt.tight_layout()
plt.savefig(file_name, dpi=300)  # Save as PNG with high resolution

# Confirm save and show the plot
print(f"Plot saved as {file_name}")
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_rel

# Data: Tabella 1 (Human)
data_human = {
    "Process": ["Connecting Objects", "Flow Objects", "Swim Lanes", "Artifacts"],
    "Ex00": [76.03, 23.97, 0.00, 0.00],
    "Ex01": [76.56, 23.44, 0.00, 0.00],
    "Ex02": [70.60, 17.18, 6.07, 6.15],
    "Ex03": [70.67, 19.34, 5.17, 4.82],
    "Ex04": [73.79, 20.24, 2.23, 3.74],
    "Ex05": [72.12, 20.84, 3.86, 3.18],
    "Ex07": [72.23, 22.97, 4.50, 0.30],
    "Ex08": [69.39, 28.84, 1.71, 0.06],
    "Ex09": [69.35, 25.19, 5.46, 0.00],
    "Ex10": [75.50, 23.55, 0.82, 0.12],
}
df_human = pd.DataFrame(data_human).set_index("Process")

# Data: Tabella 2 (LLM)
data_llm = {
    "Process": ["Connecting Objects", "Flow Objects", "Swim Lanes", "Artifacts"],
    "Ex00": [76.03, 23.97, 0.00, 0.00],
    "Ex01": [78.57, 17.86, 3.57, 0.00],
    "Ex02": [76.56, 23.44, 0.00, 0.00],
    "Ex03": [80.70, 15.79, 3.51, 0.00],
    "Ex04": [70.60, 17.18, 6.07, 6.15],
    "Ex05": [74.00, 16.00, 6.00, 4.00],
    "Ex07": [70.67, 19.34, 5.17, 4.82],
    "Ex08": [79.07, 15.12, 4.65, 1.16],
    "Ex09": [73.79, 20.24, 2.23, 3.74],
    "Ex10": [73.58, 16.98, 5.66, 3.77],
}
df_llm = pd.DataFrame(data_llm).set_index("Process")

# Calculate absolute and relative differences
absolute_diff = (df_human - df_llm).abs()
relative_diff = absolute_diff / df_human * 100

# Compute mean and standard deviation of differences for each process
stats_summary = pd.DataFrame({
    "Mean Abs Diff (%)": absolute_diff.mean(axis=1),
    "Std Abs Diff (%)": absolute_diff.std(axis=1),
    "Mean Rel Diff (%)": relative_diff.mean(axis=1),
    "Std Rel Diff (%)": relative_diff.std(axis=1),
})

# Perform paired t-test for each process
ttest_results = {}
for process in df_human.index:
    t_stat, p_value = ttest_rel(df_human.loc[process], df_llm.loc[process])
    ttest_results[process] = {"t-statistic": t_stat, "p-value": p_value}

ttest_df = pd.DataFrame(ttest_results).T

# Save and print the summary and t-test results for inspection
print("Statistical Comparison Summary:")
print(stats_summary)

print("\nPaired T-Test Results:")
print(ttest_df)

# Optionally save to CSV files for further analysis
stats_summary.to_csv("statistical_comparison_summary.csv")
ttest_df.to_csv("paired_t_test_results.csv")

# Display results
# import ace_tools as tools; tools.display_dataframe_to_user(name="Statistical Comparison Summary", dataframe=stats_summary)
# tools.display_dataframe_to_user(name="Paired T-Test Results", dataframe=ttest_df)
```
