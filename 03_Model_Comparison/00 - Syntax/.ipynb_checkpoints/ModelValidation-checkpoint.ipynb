# Python py4j conection for Model Validation

### HEPSYCODE

```python
from py4j.java_gateway import JavaGateway
import os

# Connect to the running Java EMF Validator
gateway = JavaGateway()
validator = gateway.entry_point

# Path to the metamodel (Ecore file)
metamodel_path = "C:/Users/vitto/Desktop/eclipse-modeling-2019-06-R-win32-x86_64/eclipse-modeling-2024-12-R-win32-x86_64/eclipse/HEPSYCODE/workspace/org.univaq.hepsy/model/hepsy.ecore"

# Folder that contains the .hepsy model files
models_folder = "C:/Users/vitto/Desktop/github/Re-MASTER-LLM-MODELS/02_SMG/D2_Synthetic_Model_Dataset_FS_MAS_RAG_API/D2-HEPSYCODE/LLM-ministral-8b-latest-0.0/"

# File extension to filter (without the dot)
file_extension = "hepsy"

# Initialize counters for valid and error models
valid_count = 0
error_count = 0

# Iterate over all files in the folder
for filename in os.listdir(models_folder):
    if filename.lower().endswith("." + file_extension):
        model_path = os.path.join(models_folder, filename)
        print(f"Validating {model_path}")
        result = validator.validateModel(model_path, metamodel_path, file_extension)
        print(f"\nValidation Result: {result}\n\n")
        # Check if the result contains the success message
        if "Validation successful: No errors found." in result:
            valid_count += 1
        else:
            error_count += 1

print("Validation Summary:")
print(f"Valid models: {valid_count}")
print(f"Models with errors: {error_count}")
```

```python
import os
import re
import csv
from collections import defaultdict

def classify_errors(text):
    """
    Extracts and classifies EMF validation errors from the given log text.
    Returns a dictionary: {error_type: count}
    """
    error_blocks = re.findall(
        r"Validation Result: Validation (error|errors) found:\s*(.*?)(?=\n\nValidating|\nValidation Summary|$)",
        text,
        re.DOTALL
    )

    error_counts = defaultdict(int)

    for _, block in error_blocks:
        # Look for exception types like org.eclipse.emf.ecore.xmi.FeatureNotFoundException
        exception_matches = re.findall(r"(org\.eclipse\.emf\..*?(Exception|Error))", block)
        for match in exception_matches:
            error_counts[match[0]] += 1

        # Look for feature issues like "The required feature 'id'"
        feature_matches = re.findall(r"The (required )?feature '([^']+)'", block)
        for _, feature in feature_matches:
            error_counts[f"Feature: {feature}"] += 1

        # Look for ID collisions
        if "collides with that of" in block:
            error_counts["ID Collision"] += 1

        # Catch-all for unclassified validation errors
        if not exception_matches and not feature_matches and "Validation failed" in block:
            error_counts["Unknown Validation Exception"] += 1

    return error_counts

def process_folder(folder_path, output_csv):
    """
    Processes all .txt files in the folder, extracts and classifies errors, then saves the results to a CSV.
    """
    results = []

    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            full_path = os.path.join(folder_path, filename)
            with open(full_path, 'r', encoding='utf-8') as f:
                text = f.read()
                errors = classify_errors(text)
                for error_type, count in errors.items():
                    results.append({
                        "filename": filename,
                        "error_type": error_type,
                        "count": count
                    })

    # Save to CSV
    with open(output_csv, "w", newline='', encoding='utf-8') as csvfile:
        fieldnames = ["filename", "error_type", "count"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for row in results:
            writer.writerow(row)

    print(f"CSV report saved as: {output_csv}")

# USAGE:
folder_path = "C:/Users/vitto/Desktop/github/Re-MASTER-LLM-MODELS/03_Model_Comparison/00 - Syntax/BPMN-Designer/NoRag/"
output_csv="validation_summary_BPMN-Designer_NoRAG.csv"
process_folder(folder_path,output_csv)
```
